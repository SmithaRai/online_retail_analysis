# Data Wrangling

```python
df.drop(['Email','Phone','Address','Zipcode','Product_Type','Month','State','City','Country'] , axis = 1 ,inplace = True)
```

## Data Cleaning

```python
#checking for missing values
df.isnull().mean()*100
```

```python
df.shape
```

**handling missing values**

```python
#Applying complete case analysis to handle missing values
columns = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0 or df[var].isnull().mean() == 0]
columns
```

```python
#portion of rows with null values
len(df[columns].dropna())/len(df)
```

```python
new_df = df[columns].dropna()
```

```python
df.describe()
```

```python
new_df.describe()
```

```python
new_df.info()
```

**checking for duplicates**

```python
new_df.duplicated(keep=False, subset=new_df.columns).sum()
```

```python
duplicates = new_df[new_df.duplicated(keep=False, subset=new_df.columns)]
duplicates
new_df.shape
```

```python
new_unique_df = new_df.drop_duplicates()
new_unique_df.shape
```

```python
#changing datatypes of the respective columns
new_unique_df['Amount'] = new_unique_df['Amount'].round(0).astype(int)
new_unique_df['Total_Amount'] = new_unique_df['Total_Amount'].round(0).astype(int)
new_unique_df['Total_Purchases'] = new_unique_df['Total_Purchases'].astype(int)
```

```python
new_unique_df['Date'] = pd.to_datetime(new_unique_df['Date'])
```

```python
new_unique_df['Time'] = pd.to_datetime(new_unique_df['Time'],format = '%H:%M:%S')
```

```python
#abc analysis
sorted_TotalRevenue_df = new_unique_df.sort_values('Total_Amount',ascending = False).reset_index(drop = True)

def classify(cumulative_share):
    if cumulative_share <= 70:
        return 'A'
    elif cumulative_share <= 90:  
        return 'B'
    else: 
        return 'C'

categories = []
cumulative_share = 0

for revenue in sorted_TotalRevenue_df['Total_Amount'] :
    cumulative_share = cumulative_share + revenue / sorted_TotalRevenue_df['Total_Amount'].sum()*100
    categories.append(classify(cumulative_share))

sorted_TotalRevenue_df['Category'] = categories
```

```python
sorted_TotalRevenue_df['Category'].value_counts()
```

```python
sorted_TotalRevenue_df['products'].nunique()
```

```python
# Filtering data for  category A
category_A = sorted_TotalRevenue_df[sorted_TotalRevenue_df['Category'] == 'A']
print("Category A Products:")
category_A[['Product_Category']].value_counts()

#Sorting the Category A products by Revenue in descending order
category_a_df = category_A.sort_values(by='Total_Amount', ascending=False).reset_index(drop=True)

# Calculate the number of products that make up the top 20%
top_20_percent_count = int(len(category_a_df) * 0.2)  # Get 20% of the products

# Step 4: Select the top 20% products
category_a_df['products'].head(top_20_percent_count).value_counts()

# Display the top 20% products
```
